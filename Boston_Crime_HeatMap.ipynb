```python
import pandas as pd
from sklearn.cluster import KMeans
import folium
from folium.plugins import HeatMap

# Getting the base directory 
base_directory = '/Users/nt/Desktop/DS2500/'

# Defining the file paths  
file_paths = [base_directory + f'{year}.csv' for year in range(2015, 2024)]

# Loading data 
crime_data = {str(year): pd.read_csv(path) for year, path in zip(range(2015, 2024), file_paths)}

# Get all the coordinates from 2015-2023 and count their frequencies
coordinates_freq = {}
for year in range(2015, 2024):
    if str(year) in crime_data:
        year_data = crime_data[str(year)]
        valid_coords = year_data['Location'].dropna()  
        
        # Converting data from strings to tuples and counting frequency of the coordinates 
        cords = valid_coords.apply(lambda x: eval(x))  
        for cord in cords:
            coordinates_freq[cord] = coordinates_freq.get(cord, 0) + 1 

# Converts coordinates_freq to a list of lists, so [latitude, longitude, frequency] 
graphing_data = [[cord[0], cord[1], freq] for cord, freq in coordinates_freq.items()]
   
# Computes frequency of all the streets and their assiocated districts, then combines them all
# to append onto the dictionary 
street_district_freq = {}
for year in range(2015, 2024):
    if str(year) in crime_data:
        year_data = crime_data[str(year)]
        valid_data = year_data.dropna(subset=['Location', 'STREET', 'DISTRICT'])
        for index, row in valid_data.iterrows():
            street = row['STREET']
            district = row['DISTRICT']
            street_district = (street, district)
            if street_district in street_district_freq:
                street_district_freq[street_district] += 1
            else:
                street_district_freq[street_district] = 1

# Get the top 10 most frequent streets with their district
top_10_street_district = sorted(street_district_freq, key=street_district_freq.get, reverse=True)[:10]

# Print the top 10 most dangerous streets along with their districts
print("Top 10 Most Dangerous Streets with Districts from 2015 to 2023:")
for i, (street, district) in enumerate(top_10_street_district, start=1):
    print(f"{i}. Street: {street}, District: {district}")


# Creates a map of Boston
boston = folium.Map(location=[42.3601, -71.0589], zoom_start=12)

# HeatMap using frequency as weights across all years
HeatMap(
    graphing_data,
    radius=15,
    max_zoom=13,
).add_to(boston)

# Display the map
boston
```
